# 5.Like/Unlike posts

```python
a=driver.find_element_by_xpath('/html/body/div[1]/section/nav/div[2]/div/div/div[2]/input')
a.send_keys('dilsefoodie')   
time.sleep(2)
driver.find_element_by_xpath('//*[@id="react-root"]/section/nav/div[2]/div/div/div[2]/div[3]/div/div[2]/div/div[1]/a/div').click()
```

```python
def likeUnlike(toBeLiked):
    #scroll the window
#     driver.execute_script('window.scrollTo(0, document.body.scrollHeight);')
#     time.sleep(2)

    #get the lists of posts
    posts = driver.find_elements_by_class_name("eLAPa")

    #iterate over the first 30 posts
    for i in range(30):

        #click on the post
        post = posts[i]
        post.click()
        time.sleep(2)

        #find the like button
        like = driver.find_element_by_xpath("//span[@class = 'fr66n']/button")
        svg = driver.find_element_by_css_selector(".fr66n .QBdPU svg")
        
        #conditions for clicking the like button
        if (toBeLiked==True and svg.get_attribute('aria-label') == 'Like') or (toBeLiked==False and svg.get_attribute('aria-label')=='Unlike'):
            like.click()
            
        else:
            
            #if the post is already liked
            if toBeLiked == True:
                print('post', i+1, 'is already liked')
                
            #if the post is already unliked
            else:
                print('post', i+1, 'is already unliked')
            
        

        close = driver.find_element_by_xpath("//body/div[last()]/div[last()]/button")
        close.click()
        time.sleep(2)
    

#pass True to like the posts and False to unlike the posts
# likeUnlike(toBeLiked = True)
likeUnlike(toBeLiked = False)
```

# 6.Extract List of Followers

```python
a=driver.find_element_by_xpath('/html/body/div[1]/section/nav/div[2]/div/div/div[2]/input')
a.send_keys('foodtalkindia')   
time.sleep(2)
driver.find_element_by_xpath('//*[@id="react-root"]/section/nav/div[2]/div/div/div[2]/div[3]/div/div[2]/div/div[1]/a/div').click()                                                           #first search click
wait = WebDriverWait(driver, 10)
followers_button = wait.until(EC.presence_of_element_located((By.XPATH,'//a[contains(@class,"-nal3")]'))) #follower button search
followers_button.click()                                                            #click on followers button
flag = False 
while True:
    b = wait.until(EC.presence_of_element_located((By.CLASS_NAME,'isgrP')))         #search for followers
    p = driver.find_element_by_class_name('isgrP')
    
    p.click()                                                        
    driver.find_element_by_tag_name('body').send_keys(Keys.END)                    #moving end of document 
    if flag == False:
        for i in range(6):
            driver.find_element_by_tag_name('body').send_keys(Keys.PAGE_UP)        #to remove suggestion buttuon 
            time.sleep(1) 
        flag = True                                                                #only work for one time
    time.sleep(1)
    p = driver.find_element_by_class_name('isgrP')                                 #click on page of follower
    p.click()
    follower_list = driver.find_elements_by_xpath('//div[@class = "isgrP"]/ul/div/li')  #follower list
    if len(follower_list) >= 500:                                                       #cheking for only 500 follower
        break
li = []
j=0
foodtalkindia_follower = []
for i in follower_list:                                                                  
    data = BeautifulSoup(i.get_attribute('innerHTML'),'html')                         #to get the name of follower             
    li.append(data.a['href'].split('/')[1])                                           #adding name in li
    foodtalkindia_follower.append(data.a['href'].split('/')[1])
    j+=1
    if j==500:                                                                        #only top 500 follower
        break
for i in foodtalkindia_follower:
    print(i)
```

```python
driver.back()
```

# 7.Check the story of ‘coding.ninjas’. Consider the following Scenarios and print error messages accordingly -
##    1.If You have already seen the story.
##    2.Or The user has no story.
##    3.Or View the story if not yet seen.

```python
a=driver.find_element_by_xpath('/html/body/div[1]/section/nav/div[2]/div/div/div[2]/input')
a.send_keys('coding ninjas')   
time.sleep(2)
driver.find_element_by_xpath('//*[@id="react-root"]/section/nav/div[2]/div/div/div[2]/div[3]/div/div[2]/div/div[1]/a/div').click() 
```

Here I have observed that the handles which have story have class h5uC0.The stories which are already seen have hight and width-208 while which have not yet been seen have height 210 and width 210.

```python
time.sleep(5)
l= driver.find_elements_by_class_name("h5uC0")
s = len(l)
if(s>0):
    Height = BeautifulSoup(driver.find_element_by_xpath("//div[@class = 'XjzKX']//canvas").get_attribute("outerHTML"),'html.parser').canvas['height']
    Width = BeautifulSoup(driver.find_element_by_xpath("//div[@class = 'XjzKX']//canvas").get_attribute("outerHTML"),'html.parser').canvas['width']
    if int(Height) == 208 and int(Width) == 208:
        print("Already seen the story.")
    else:
        print("View the story if not yet seen.")
        a = wait.until(EC.presence_of_element_located((By.CLASS_NAME,'_2dbep')))              
        a.click() 
        print("Seen now")
else:
    print("The user has no story")      
```
